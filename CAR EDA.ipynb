```python
'''Remove non-numeric characters and convert kmDriven to numeric'''
data['kmDriven'] = data['kmDriven'].str.replace('[^0-9]', '',regex=True).astype(float)


'''Remove non-numeric characters and convert ASKPrice to numeric'''
data['AskPrice'] = data['AskPrice'].str.replace('[^0-9]', '',regex=True).astype(float)


'''Convert postedDate to a standdard datetime format'''
data['PostedDate'] = pd.to_datetime(data['PostedDate'], format='%b-%y',errors='coerce').dt.date
```

```python
data.info()
```

```python
'''Impute missing values in kmDriven with the median value'''
median_kmDriven = data['kmDriven'].median()
data.fillna({'kmDriven': median_kmDriven},inplace=True)

```

```python
data.isnull().sum()
```

```python
numerical_columns = data[['Year', 'Age','kmDriven','AskPrice']].columns
categorical_columns = data[['Brand','Transmission','Owner','FuelType']].columns
```

```python
numerical_columns
```

```python
''' Univariate Analysis
Objective: To explore each variable independently.
Insights and 
Examples:
Histograms for numerical variables like price or mileage, highlighting skewness or common ranges.
Boxplots showing outliers in variables like engine size or car price.
Countplots for categorical data such as fuel type or car brand, showing frequency distribution.'''
#Univariate Analysis for Numerical Variables
#Plotting histograms
for column in numerical_columns:
    plt.figure(figsize=(8,4))
    sns.histplot(data[column],kde=True, bins=30, color='Olive')
    plt.title(f'Distribution of {column}', fontsize=14)
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.show()
```

```python
'''outlier identification'''
plt.figure(figsize=(10, 10))


for idx, col in enumerate(numerical_columns):
    plt.subplot(2, 2, idx + 1)
    sns.boxplot(x=col, data=data)
    plt.title(col)

plt.show()
    
```

```python



#Univariate Analysis for Categorical Variables

categorical_columns = ['Brand','Transmission','Owner','FuelType']
hue_column = 'Transmission'


#plotting bar plots with hue
for column in categorical_columns:
    plt.figure(figsize=(12, 6))
    sns.countplot(x=data[column],hue=data[hue_column], order=data[column].value_counts().index, palette='viridis')
    plt.title(f'Count of {column} by {hue_column}', fontsize=14)
    plt.xlabel(column)
    plt.ylabel('Count')
    plt.xticks(rotation=45)
    plt.legend(title=hue_column)
    plt.show()
```

```python
''' Bivariate Analysis
Objective: To examine relationships between two variables.
Insights and Examples:
Scatterplot of price vs. mileage, showing an inverse relationship.
Boxplot comparing price by fuel type, indicating that diesel cars are generally priced higher.
Correlation heatmap showing the strength of linear relationships between numerical variables.'''
# bivariate Analysis for Numerical vs. Numerical

import matplotlib.pyplot as plt
import seaborn as sns

#Scatter plot with Age vs. ASKprice, and hue based on Transmission

plt.figure(figsize=(8, 5))
sns.scatterplot(x='Age', y='AskPrice', data=data, hue='Transmission')
plt.title('Age vs. AskPrice')
plt.xlabel('Age (Years)')
plt.ylabel('AskPrice')
plt.show()
```
